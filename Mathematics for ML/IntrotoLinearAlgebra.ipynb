Inverse limitations:
1. Matrix should not be singular(all columns of the matrix should be linearly independent)
2. Matrix should be a square matrix. (#rows = #columns i.e., "Vector span = Matrix range")
   . Matrix is over-determined (#rows> #columns)
   . Matrix is under - dertermined (#rows < #columns)

Traces:

1. Sum of all the diagonal elements of a matrix
2. Tr(A) = Tr(A.T) , Transpose of a matrix and the matrix have same trace value
3. Multiplication of 3 matrices in any order gives the same trace value.
4. to calculate the forbenius norm of a matrix ||A|| = sqrt(tr(AA.T))

```python
A_p = torch.tensor([[-1,2],[3,-2],[5,7.]])
```

```python
trace_Ap = torch.trace(A_p)
trace_Ap
```

```python
Anorm_p = torch.linalg.norm(A_p)
Anorm_p
```

```python
import math
```
